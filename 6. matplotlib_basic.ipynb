# Matplotlib 연습장
데이터 시각화를 위한 기본적인 matplotlib 예제를 알아봅시다.

## 1. Matplotlib 사용하기
소스코드 최상단에 아래와 같이 입력합니다. 보통 plt를 약자로 사용합니다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. 기본 그래프 그리기
`plot()`함수를 이용해 대부분의 그림을 그릴 수 있습니다.
  
우선 간단한 그래프를 그려보겠습니다.

```python
x = np.arange(0,10,1)
x
```

```python
y = np.random.randn(10)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
y1 = np.random.randn(10)
y2 = np.random.randn(10)

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

## 그래프 옵션 조정
1. `plot()` 옵션
- `color` : 선 색깔 변경
- `marker` : 데이터 꺾임 부분 표시
- `label` : 범례 지정
- `linestyle` : 선 모양

1. 그래프 옵션
- `plt.xlabel()` : x축 설명
- `plt.ylabel()` : y축 설명
- `plt.title()` : 그래프 제목
- `legend()` : 범례 표시
- `plt.xlim()` : x축 범위
- `plt.ylim()` : y축 범위

```python
plt.plot(x, y1, color='red', marker='.')
plt.plot(x, y2, color='black', marker='*', linestyle='--')
plt.show()
```

Google에 matplotlib => document, tutorial 등 검색하면 잘 정리되어 있음

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color='red', linewidth=0.25)
ax.plot(x, x+2, color='red', linewidth=0.50)
ax.plot(x, x+3, color='red', linewidth=1.00)
ax.plot(x, x+4, color='red', linewidth=2.00)

# 라인스타일 옵션 '-', '-.', ':', 'steps'
ax.plot(x, x+5, color='green', lw=3, linestyle='-')
ax.plot(x, x+6, color='green', lw=3, ls='-.')
ax.plot(x, x+7, color='green', lw=3, ls=':')

# 임의의 선 모양
line, = ax.plot(x, x+8, color='black', lw=1.50)
line.set_dashes([5,10,15,10])

# 마커 유형 '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4',...
ax.plot(x, x+9, color='blue', lw=3, ls='-', marker='+')
ax.plot(x, x+10, color='blue', lw=3, ls='--', marker='o')
ax.plot(x, x+11, color='blue', lw=3, ls='-', marker='s')
ax.plot(x, x+12, color='blue', lw=3, ls='--', marker='1')

# 마커 사이즈 및 색상
ax.plot(x, x+13, color='purple', lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color='purple', lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color='purple', lw=1, ls='-', marker='o', markersize=8, markerfacecolor='red')
ax.plot(x, x+16, color='purple', lw=1, ls='-', marker='s', markersize=8, markerfacecolor='yellow',
       markeredgewidth=3, markeredgecolor='green')
```

## 4. 빈 그림판을 만든 후 내용을 채우는 방법

```python
# 빈 그림판 생성
fig = plt.figure()

# 그림 크기 설정
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y, 'b')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

## 5. 그림 크기와 화소 설정
- figsize : 그림 크기
- dpi : 인치 당 픽셀 개수

```python
fig = plt.figure(figsize=(10,4), dpi=100)

axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

## 6. 그림 저장하기
- `fig.savefig()`

```python
fig.savefig('filename.png')
```

## 7. 다양한 그래프 형식
- 산점도 그래프 : scatter
- 히스토그램 : hist
- 상자수염 그래프 : boxplot

```python
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range(1,1000), 100)
plt.hist(data)
```

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]
plt.boxplot(data, vert=True, patch_artist=True)
```
