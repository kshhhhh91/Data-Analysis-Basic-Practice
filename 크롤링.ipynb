```python
# 이제 for문을 이용하여 pandas로 데이터프레임으로 만들수 있음.
```

```python
title_list = []
content_list = []
name_list = []
date_list = []

for i in range(len(soup.select('#news_area > .news_list'))):
    title_list.append(soup.select('#news_area > .news_list')[i]('span')[0].text)
    content_list.append(soup.select('#news_area > .news_list')[i]('a')[1].text)
    name_list.append(soup.select('#news_area > .news_list')[i]('span')[1].text.split(" | ")[0])
    date_list.append(soup.select('#news_area > .news_list')[i]('span')[1].text.split(" | ")[1])
    
df = pd.DataFrame(list(zip(title_list, content_list, name_list, date_list)), columns=['제목', '내용', '기자', '날짜'])
df
```

```python
title = []
content = []
writer = []
date = []

for i in range(1,31):
    url = 'http://www.boannews.com/media/t_list.asp?Page=' + str(i)
    request = Request(url)
    response = urlopen(request)
    html = response.read()
    soup = BeautifulSoup(html, 'html.parser')   # 함수로 묶어도 됨
    
    for j in soup.select('#news_area > .news_list'):
        title.append(j('span')[0].text)
        content.append(j('a')[1].text)
        writer.append(j('span')[1].text.split(" | ")[0])
        date.append(j('span')[1].text.split(" | ")[1])

df2 = pd.DataFrame(list(zip(title,content,writer,date)), columns=['제목','내용','기자','날짜'])
df2
```

```python
df2.to_csv('news_crawling1.csv', index=False)
```
