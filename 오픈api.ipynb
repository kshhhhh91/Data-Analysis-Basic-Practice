```python
# open_api 주소 정보
# 서비스키
# 목록에서 한번에 몇개씩 요청 될 지
# 페이지 넘버
# 시,도
# 시,군,구
# 건물명 (필요없음)
```

```python
url = 'http://apis.data.go.kr/openapi/service/ElevatorInformationService/getElevatorList?'
serviceKey='D%2BNlB5fEAMk%2BJbNvi2wxyeTG4iAXa1%2B2baF8Yd7OobnnqdJvGfXHC2LdgdcQhVtXlQSxG1cvPwcRWUlQoCgA%2Bg%3D%3D'
numOfRows=100
pageNo=1
sido='%EC%84%9C%EC%9A%B8%ED%8A%B9%EB%B3%84%EC%8B%9C'
sigungu='%EC%9D%80%ED%8F%89%EA%B5%AC'

api_url = url + 'serviceKey='+serviceKey + '&numOfRows='+\
            str(numOfRows) + '&pageNo=' + str(pageNo) +\
            '&sido=' + sido + '&sigungu=' + sigungu

# \는 줄바꿈용
api_url
```

#### (3) 가져온 데이터 파싱하기

- 승강기 주소1 (address1)
- 승강기 주소2 (address2)
- 건물명 (buldNm)
- 적재하중  (liveLoad)
- 제조업체 전화번호  (manufacturerTelno)

```python
res = requests.get(api_url)
```

```python
xml = res.text   # open api는 xml 형태로 파싱
soup = BeautifulSoup(xml, 'xml')
soup
```

```python
soup.select('address1')
```

```python
soup.select('address2')[3].text
```

```python
len(soup.select('manufacturerTelno'))
```

```python
add1 = []
add2 = []
buld = []
live = []
manu = []

for i in soup.select('address1'):
    add1.append(i.text)
for i in soup.select('address2'):
    add2.append(i.text)
for i in soup.select('buldNm'):
    buld.append(i.text)
for i in soup.select('liveLoad'):
    live.append(i.text)
for i in soup.select('manufacturerTelno'):
    manu.append(i.text)
                
df = pd.DataFrame(list(zip(add1, add2, buld, live, manu)), columns = ['승강기주소1', '승강기주소2', '건물명', '적재하중', '제조업체 전화번호'])
df
```

```python
# 99개 밖에 안옴. 제조업체 전화번호가 하나 빠짐. 결측치가 있는 행을 날려버림. list(zip( 을 해서 그렇게 됨. 다른 방식이 필요함.
```

```python
add1 = []
add2 = []
buld = []
live = []
manu = []

for i in soup.select('address1'):
    add1.append(i.text)
for i in soup.select('address2'):
    add2.append(i.text)
for i in soup.select('buldNm'):
    buld.append(i.text)
for i in soup.select('liveLoad'):
    live.append(i.text)
for i in soup.select('manufacturerTelno'):
    manu.append(i.text)
    
df_add1 = pd.DataFrame(add1, columns=['주소1'])
df_add2 = pd.DataFrame(add2, columns=['주소2'])
df_buld = pd.DataFrame(buld, columns=['건물명'])
df_live = pd.DataFrame(live, columns=['적재하중'])
df_manu = pd.DataFrame(manu, columns=['연락처'])
df = pd.concat([df_add1, df_add2, df_buld, df_live, df_manu], axis=1)
df
```

```python
# 이 데이터프레임의 문제는, 전화번호의 결측치가 그냥 맨 마지막 행으로 가고, 전화번호가 하나씩 밀려있다는 것.
```
