#### 고객들이 직업별로 구매금액이 어떻게 될까요?

```python
df_tmp = df[['Job', 'Purchase Price']]
df_tmp
```

```python
df_tmp.groupby('Job')  # 직업 기준으로 그룹을 묶어주는 object 생성
```

```python
df_tmp.groupby('Job').count()  # 직업 별 구매 건수
```

```python
df_tmp.groupby('Job').count().sort_values(by='Purchase Price')
```

```python
df_tmp.groupby('Job').mean()  # 직업 별 구매금액 평균
```

#### 어떤 직업을 가진 사람들이 가장 많은 직업을 지출했나요?

```python
df_tmp.groupby('Job').sum()
```

```python
df_tmp.groupby('Job').sum().sort_values(by='Purchase Price')
```
